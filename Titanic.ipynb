```python
import pandas as pd
```

```python
tit=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
tit.head(15)
```

```python
tit.isnull().sum()
```

```python
tit.Age.fillna(tit.Age.mean(),inplace=True)
```

```python
tit.Age.isnull().sum()
```

```python
tit.Cabin.fillna(tit.Cabin.mode(),inplace=True)
```

```python
tit.Cabin.isnull().sum()
```

```python
tit.Embarked.fillna(tit.Embarked.mode(),inplace=True)
```

```python
tit.Embarked.isnull().sum()
```

```python
tit.drop(['Name','PassengerId','Cabin','Ticket','Embarked'],axis=1)
```

```python
x=tit[['Age','Fare']]
y=tit.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(tit,test_size=20)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression().fit(train_x,train_y)
model.score(train_x,train_y)
```

```python
predict_y=model.predict(test_x)
```

```python
model.predict_proba(test_x)
```

```python
#confusion matrix
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(test_y,predict_y)
```
